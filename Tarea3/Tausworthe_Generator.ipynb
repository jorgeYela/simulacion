```python
r = 3
q = 5
l = 4
p = (2 ** q) - 1
```

```python
B = {}
```

```python
for i in range(1,q+1):
    key = "B" + str(i)
    B[key] = 1
```

```python
print(B)
```

```python
def xor(br,bq):
    new_xor = 0
    if br == bq:
        new_xor = 0
    else:
        new_xor = 1
    return new_xor
```

```python
for i in range(1,p-r):
    keyo = "B" + str(i + q)
    keyr = "B" + str((i + q) - r)
    keyq = "B" + str((i + q) - q)
    B[keyo] = xor(B.get(keyr), B.get(keyq))
```

```python
print(B)
```

```python
binarios = []
for key in B:
    binarios.append(B[key])
```

```python
n = 4
output = [binarios[i:i + n] for i in range(0, len(binarios), n)]
```

```python
binarios_generados = []
for i in range(0,len(output)):
    binarios_generados.append("".join(str(_)for _ in output[i]))
```

```python
print(binarios_generados)
```

```python
def binario_a_decimal(binario):
    posicion = 0
    decimal = 0
    binario = binario[::-1]
    for digito in binario:
        multiplicador = 2 ** posicion
        decimal += int(digito) * multiplicador
        posicion += 1
    return decimal
```

```python
for binario in binarios_generados:
    aleatorio = binario_a_decimal(binario) / (2 ** l)
    print(aleatorio)
```
