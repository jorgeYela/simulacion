```python
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np
```

## 1. Distribución Pareto

### 1.1 Introducción:

La distribución pareto viene dada por la siguiente formula: 

$$F(x) = Pr(X > x) = \frac{\left(\begin{array}{c} Xm \\ x \end{array}\right)^α ; x >= xm}{1 ; x < xm}
$$

Donde: 

a) Xm es el valor mínimo posible > 0

b) α es el índice de Pareto > 0 (Parametro de forma)

### 1.2 Descripción:

La distribución de pareto nos permite conocer el orden de imoprtancia de las variables que intervienen en nuestro experimento. Esta distribución muestra que el 80% de las variables estudiadas no cuentan con un alto grado de relevancia, mientras ue el 20% restante es el que realmente interesa en el estudio. 

Un ejemplo de implementación esta ligado al marketing, donde aproximandamente solamente el 20% de los clientes de determinada empresa generan el 80% de ingresos economicos, esto nos indica que es a ese 20% de clientes es al que tiene que estar centrada la atención empresarial.


### 1.3 Ejemplo de uso:

```python
# PARAMETROS DE FORMA (ÍNDICE DE PARETO)
b = 2.62

# DISTRIBUCIÓN PARETO
x = np.linspace(stats.pareto.ppf(0.01, b), stats.pareto.ppf(0.99, b), 100)

# FUNCIÓN DE DENSIDAD PROBABILIDAD
fp = stats.pareto.pdf(x, b)

# GRÁFICO
fig, ax = plt.subplots()
ax.plot(x, fp, '--')
ax.vlines(x, 0, fp, colors='b', lw=5, alpha=0.5)
plt.title('Distribución Pareto')
plt.ylabel('Probabilidad')
plt.xlabel('Valores')
plt.show()
```

```python
# GENERACIÓN DE ALEATORIOS
aleatorios = stats.pareto.rvs(b, scale=1, size=100)
cuenta, cajas, ignorar = plt.hist(aleatorios, 20)

# GRÁFICA HISTOGRAMA
plt.ylabel('Frecuencia')
plt.xlabel('Valores')
plt.title('Histograma Pareto')
plt.show()
```

```python
print(aleatorios)
```

## 2. T de Student

### 2.1 Introducción:

La distribución viene dada por la siguiente función de densidad:

$$h(t) = \frac{r\left(\begin{array}{c} k + 1 \\ 2 \end{array}\right)}{r\left(\begin{array}{c} k \\ 2 \end{array}\right)\sqrt{πk}}\left(1+\begin{array}{c} t^2 \\ k \end{array}\right)^{-\frac{(k+1)}{2}}$$

Donde:

a) k grados de libertad > 0

b) t es evaluada en -∞ < t < ∞

### 2.2 Descripción:

La distribución de t de Student permite estimar la media de una población normalmente distribuida cuando el tamaño del universo es relativamente pequeño.

La curva de la distribución t de Student tiene forma de campana con centro en 0, su curva esta mas dispersa que la curva normal estándar, si los grados de libertad incrementan la dispersión de la curva disminuye, cuanto mas al ∞ tiendan los grados de libertad la curva se aproxima a una curva normal estándar.

Un ejemplo de uso se da cuando se desea calibrar determinada herramienta de medición, para esto se especifica un valor medio y los grados de libertad para que despues el instrumento pase a ser medido en 10 ocaciones con el objetivo de que los resultados se acerquen o aproximen a la media. 

### 2.3 Ejemplo de uso:

```python
# PARAMETROS DE FORMA
df = 2.74

# DISTRIBUCIÓN PARETO
x = np.linspace(stats.t.ppf(0.01, df), stats.t.ppf(0.99, df), 100)

# FUNCIÓN DE DENSIDAD PROBABILIDAD
fp = stats.t.pdf(x, df)

# GRÁFICO
fig, ax = plt.subplots()
ax.plot(x, fp, '--')
ax.vlines(x, 0, fp, colors='b', lw=5, alpha=0.5)
ax.set_yticks([0., 0.1, 0.2, 0.3, 0.4])
plt.title('Distribución T Student')
plt.ylabel('Probabilidad')
plt.xlabel('Valores')
plt.show()
```

```python
# GENERACIÓN DE ALEATORIOS
aleatorios = stats.t.rvs(df, size=100)
cuenta, cajas, ignorar = plt.hist(aleatorios, 20)

# GRÁFICA HISTOGRAMA
plt.ylabel('Frecuencia')
plt.xlabel('Valores')
plt.title('Histograma T Student')
plt.show()
```

```python
print(aleatorios)
```

## 3. Chi Cuadrado

### 3.1. Introducción:

La distribución viene dada por la siguiente función de densidad:

$$f(x, k) = \frac{1}{2^{k/2}r(k/2)}x^{k/2-1}exp(-x/2)$$

Donde:

a) x > 0

b) k > 0

### 3.2 Descripción:

La distribución de chi cuadrado esta dada según los grados de libertad y el parametro de no centralidad. La distribución es positivamente asimétrica, esta disminuye cuando los grados de libertad aumenta. 

Es utilizada mayoritariamente para verificar que tan bien ajustada se encuentra determinada muestra. Un ejemplo de uso es el algoritmo de chi cuadrado que permite realizar una prueba de bondad de ajuste que determina si los datos de una muestra aleatoria se ajusta a una distribución uniforme. 

### 3.3 Ejemplo de uso:

```python
# PARAMETROS DE FORMA
df = 55

# DISTRIBUCIÓN PARETO
x = np.linspace(stats.chi2.ppf(0.01, df), stats.chi2.ppf(0.99, df), 100)

# FUNCIÓN DE DENSIDAD PROBABILIDAD
fp = stats.chi2.pdf(x, df)

# GRÁFICO
fig, ax = plt.subplots()
ax.plot(x, fp, '--')
ax.vlines(x, 0, fp, colors='b', lw=5, alpha=0.5)
plt.title('Distribución Chi Cuadrado')
plt.ylabel('Probabilidad')
plt.xlabel('Valores')
plt.show()
```

```python
# GENERACIÓN DE ALEATORIOS
aleatorios = stats.chi2.rvs(df, size=100)
cuenta, cajas, ignorar = plt.hist(aleatorios, 20)

# GRÁFICA HISTOGRAMA
plt.ylabel('Frecuencia')
plt.xlabel('Valores')
plt.title('Histograma Chi Cuadrado')
plt.show()
```

```python
print(aleatorios)
```

## 4. Beta

### 4.1 Introducción:

La distribución viene dada por la siguiente función de densidad:

$$f(x, a, b) = \frac{r(a+b)x^{a-1}(1-x)^{b-1}}{r(a)r(b)}$$

Donde:

a) 0 <= x <=1

b) a > 0

c) b > 0

d) r es la función gamma

### 4.2 Descripción:

La distribución beta es utilizada normalmente para representar la variabilidad dentro de determinado rango, es utilizada también para describir datos empíricos y predecir el comportamiento aleatorio con el fín de, en un caso de uso, determinar el grado de fiabilidad de los dispositivos informáticos de determinada empresa.

### 4.3 Ejemplo de uso:

```python
# PARAMETROS DE FORMA
a, b = 2.31, 0.627

# DISTRIBUCIÓN PARETO
x = np.linspace(stats.beta.ppf(0.01, a, b), stats.beta.ppf(0.99, a, b), 100)

# FUNCIÓN DE DENSIDAD PROBABILIDAD
fp = stats.beta.pdf(x, a, b)

# GRÁFICO
fig, ax = plt.subplots()
ax.plot(x, fp, '--')
ax.vlines(x, 0, fp, colors='b', lw=5, alpha=0.5)
plt.title('Distribución Beta')
plt.ylabel('Probabilidad')
plt.xlabel('Valores')
plt.show()
```

```python
# GENERACIÓN DE ALEATORIOS
aleatorios = stats.beta.rvs(a, b, size=100)
cuenta, cajas, ignorar = plt.hist(aleatorios, 20)

# GRÁFICA HISTOGRAMA
plt.ylabel('Frecuencia')
plt.xlabel('Valores')
plt.title('Histograma Chi Cuadrado')
plt.show()
```

```python
print(aleatorios)
```

## 5. Gamma

### 5.1 Introducción:

La distribución gamma viene dada por la siguiente función de probabilidad:

$$f(x, a) = \frac{x^{a-1}e^{-x}}{r(a)}$$

Donde:

a) x >= 0

b) a > 0

c) r se refiere a la función gamma

### 5.2 Descripción:

La distribución gamma modela valores de datos positivos que sean asimétricos a la derecha y mayores que 0. La distribución gamma es una distribución continua que se define por sus parámetros de forma y escala. 

En un caso practico la distribución gamma puede puede describir el tiempo que transcurre para que determinado comonente eléctrico falle, la mayoria de estos componentes fallaran en un periodo similar de tiempo pero algunos tardarán o se apresuraran a fallar con respecto a la mayoria.

### 5.3 Ejemplo de uso:

```python
# PARAMETROS DE FORMA
a = 1.99

# DISTRIBUCIÓN PARETO
x = np.linspace(stats.gamma.ppf(0.01, a), stats.gamma.ppf(0.99, a), 100)

# FUNCIÓN DE DENSIDAD PROBABILIDAD
fp = stats.gamma.pdf(x, a)

# GRÁFICO
fig, ax = plt.subplots()
ax.plot(x, fp, '--')
ax.vlines(x, 0, fp, colors='b', lw=5, alpha=0.5)
plt.title('Distribución Gamma')
plt.ylabel('Probabilidad')
plt.xlabel('Valores')
plt.show()
```

```python
# GENERACIÓN DE ALEATORIOS
aleatorios = stats.gamma.rvs(a, size=100)
cuenta, cajas, ignorar = plt.hist(aleatorios, 20)

# GRÁFICA HISTOGRAMA
plt.ylabel('Frecuencia')
plt.xlabel('Valores')
plt.title('Histograma Gamma')
plt.show()
```

```python
print(aleatorios)
```

## 6. Log Normal

### 6.1 Introducción:

La distribución lognormal viene dada por la siguiente ecuación de densidad:

$$f(x, s)=\frac{1}{sx\sqrt{2π}}exp\left(-\begin{array}{c} log^2(x) \\ 2s^2 \end{array}\right)$$

Donde:

a) x > 0

b) s > 0

### 6.2 Descripción:

La distribución lognormal es una distribución continua que se define por sus parámetros de ubicación y escala.

En un caso de uso, la distribución lognormal permite analisar la fiabilidad de crecimiento financiero y modelar el comportamiento de acciones economicas.

### 6.3 Ejemplo de uso:

```python
# PARAMETROS DE FORMA
s = 0.954

# DISTRIBUCIÓN PARETO
x = np.linspace(stats.lognorm.ppf(0.01, s), stats.lognorm.ppf(0.99, s), 100)

# FUNCIÓN DE DENSIDAD PROBABILIDAD
fp = stats.lognorm.pdf(x, s)

# GRÁFICO
fig, ax = plt.subplots()
ax.plot(x, fp, '--')
ax.vlines(x, 0, fp, colors='b', lw=5, alpha=0.5)
plt.title('Distribución Log Normal')
plt.ylabel('Probabilidad')
plt.xlabel('Valores')
plt.show()
```

```python
# GENERACIÓN DE ALEATORIOS
aleatorios = stats.lognorm.rvs(s, size=100)
cuenta, cajas, ignorar = plt.hist(aleatorios, 20)

# GRÁFICA HISTOGRAMA
plt.ylabel('Frecuencia')
plt.xlabel('Valores')
plt.title('Histograma Log Normal')
plt.show()
```

```python
print(aleatorios)
```
