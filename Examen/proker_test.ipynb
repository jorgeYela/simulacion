```python
import random

# FUNCIÓN GENERADORA 100 NÚMEROS PSEUDOALEATORIOS PYTHON RANDOM
def random_python(N=100):
    numeros = []
    random.seed(1032022)
    [numeros.append(random.random())for i in range(N)]
    return numeros
```

```python
# FUNCIÓN GENERADORA 100 NÚMEROS PSEUDOALEATORIOS CUADRADOS MEDIOS
def cuadradosMedios(xo, digitos):
    iteraciones = 100
    xn = xo
    aleatorios=[]
    for i in range(iteraciones):
        xnn = xn**2
        txnn = str(xnn).zfill(8)
        tam = len(txnn)
        ui = int(txnn[int(tam/2-digitos/2):int(tam/2+digitos/2)])
        rn = ui / (int('9'*digitos)+1)
        rn=str(rn)
        aleatorios.append(rn)
        xn = ui
    return(aleatorios)
```

```python
# FUNCIÓN GEERADORA 100 NÚMEROS PSEUDOALEATORIOS CONGRUENCIA LINEAL
def congruencia_lineal(a,b,xo,m):
    iteraciones = 100
    xn = xo
    aleatorios=[]
    for i in range(iteraciones):
        xnn = (a*xn+b) % m
        ui = xnn/m
        ui=str(ui)
        aleatorios.append(ui)
        xn = xnn
    return aleatorios
```

```python
from collections import Counter

def conteo_poker(cad_numeros):
  result = []
  for mano in cad_numeros:
    conteo=Counter(mano)
    resultado={}
    for clave in conteo:  
      valor=conteo[clave]
      if valor != 1:
        resultado[clave] = valor
    result.append(resultado)
  return result
```

```python
def par(dici):
  cadena_resultados = []
  for i in range(len(dici)):
    a = list(dici[i].values())
    for j in range(len(a)):
      if(a[j] == 2):
        cadena_resultados.append("1P")
      elif(a[j] == 3):
        cadena_resultados.append("T")
      elif(a[j] == 2) and (a[j+1] == 2):
        cadena_resultados.append("2P")
      elif(a[j] == 3) and (a[j+1] == 2):
        cadena_resultados.append("TP")
      elif(a[j] == 4):
        cadena_resultados.append("P")
      elif(a[j] >= 5):
        cadena_resultados.append("Q")
```

```python
def quintilla(numero):
    digito1 = numero[0]
    for digito in numero:
        if digito != digito1:
            return False
    return True
```

```python
def full(numero):
    guia = dict.fromkeys(numero, 0)
    for digito in numero:
        guia[digito]+=1
    if(2 in guia.values() and 3 in guia.values()):
        return True
    return False
```

```python
def tercia(numero):
    # Conteo
    guia = dict.fromkeys(numero, 0)
    for digito in numero:
        guia[digito]+=1
    # Impar
    for conteo in guia.values():
        if conteo >= 3:
            return True
    return False
```

```python
def poker(numero):
    if(tercia(numero)):
        guia = dict.fromkeys(numero, 0)
        for digito in numero:
            guia[digito]+=1
        for conteo in guia.values():
            if conteo >= 4:
                return True
        return False
    else:
        return False
```

```python
def onep(numero):
    # Conteo
    guia = dict.fromkeys(numero, 0)
    for digito in numero:
        guia[digito]+=1
    # Par
    for conteo in guia.values():
        if conteo >= 2:
            return True
    return False
```

```python
def twop(numero):
    guia = dict.fromkeys(numero, 0)
    for digito in numero:
        guia[digito]+=1
    if onep(numero):
        par = None
        for conteo in guia.items():
            if conteo[1] >= 2:
                par = conteo[0]
                break
        del guia[par]
        for conteo in guia.values():
            if conteo >= 2:
                return True
        return False
    else:
        return False
```

```python
def td(numero):
    return not (len(numero) != len(set(numero)))
```

```python
def categorizacion(numeros):
  categorias = []
  for numero in numeros:
    N = len(numero)
    if N != 1:
      if quintilla(numero):
        categorias.append('Q')
      elif poker(numero):
          categorias.append('P')
      elif full(numero):
          categorias.append('TP')
      elif tercia(numero):
          categorias.append('T')
      elif twop(numero):
          categorias.append('2P')
      elif onep(numero):
          categorias.append('1P')
      else:
          categorias.append('TD')
    else:
      categorias.append('TD')
  return categorias
```
