# Prueba Unidad 3 - Método de la Transformada Inversa

## 1. Problema

El tiempo de preparación de determinado pedido en una cadena de comida rápida viene dado por una media de 5 minutos por pedido, cabe recalcar que dichos tiempos se encuentra distribuidos exponencialmente. Simular el comportamiento de la variable aleatoria mediante el método de la transformada inversa para la mencionada <strong>"distribución exponencial"</string>

## 2. Simulación

```python
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np
import math
```

```python
import random

# GENERACIÓN DE NÚMEROS ALEATORIOS
def random_python(N=100):
    numeros = []
    random.seed(1032022)
    [numeros.append(random.random())for i in range(N)]
    return numeros
```

```python
# DISTRIBUCIÓN EXPONENCIAL
exponencial = stats.expon()
```

```python
# FUNCIÓN INVERSA
def inversa(aleatorios,λ):
  tiempos = []
  for aleatorio in aleatorios:
    xi = abs(-(math.log((abs(1.0-aleatorio))))/(1/λ))
    tiempos.append(xi)
  return tiempos
```

```python
# FUNCIÓN PARA PRESENTAR RESULTADOS CON N NUMERO DE DECIMALES
def truncate(num, n):
  integer = int(num * (10**n))/(10**n)
  return float(integer)
```

```python
# 100 ALEATORIOS GENERADOS Y VALOR DE LA MEDIA

λ = 5
ri = random_python(100)
# ri = exponencial.rvs(100)
# ri = [0.64,0.83,0.03,0.50,0.21]

inverso = inversa(ri, λ)
```

## 3. Resultados

### 3.1 Comportamiento de las variables aleatorias

```python
print("Resultados Obtenidos:\n")
print("Pedido Nº \t ri \t tiempo")
for i in range(len(ri)):
  print("Pedido " + str(i+1) + "\t" + str(truncate(ri[i],2)) + "\t" + str(truncate(inverso[i],2))+" minutos")
```

### 3.2 Gráfico de la Función de Densdad

```python
fp = exponencial.pdf(ri)

plt.plot(ri,fp,"o")
plt.title("Función Densidad")
plt.ylabel("F(x)")
plt.xlabel("x")
plt.show()
```

### 3.3 Gráfico de la Función de Probabilidad Acumulada

```python
cdf = exponencial.cdf(inverso)

plt.plot(inverso,cdf,"o")
plt.title("Función Probabilidad Acumulada")
plt.ylabel("F(x)")
plt.xlabel("x")
plt.show()
```
