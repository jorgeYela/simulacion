## 1. Conjunto de Datos FIFA

```python
import pandas as pd
import io

data = pd.read_csv(io.StringIO(uploaded['fifa_datos.csv'].decode('utf-8')))

data.head(5)
```

```python
from scipy import stats
import pandas as pd
import numpy as np
```

```python
mAritmetica_fila_fifa = data.mean(axis=1)

print("Media Aritmetica Por Fila: \n" + str(mAritmetica_fila_fifa))
```

```python
mAritmetica_columna_fifa = data.mean(axis=0)

print("Media Aritmetica Por Columna: \n" + str(mAritmetica_columna_fifa))
```

```python
values = data["Value"]
wages = data["Wage"]

intValues = []
for value in values:
  a = value.replace("M","000000")
  b = a.replace("K","000")
  c = b.replace("€","")
  intValues.append(float(c))

intWages = []
for wage in wages:
  a = wage.replace("M","000000")
  b = a.replace("K","000")
  c = b.replace("€","")
  intWages.append(float(c))

print("Media Overrall: " + str(np.median(data["Overall"])))
print("Media Potential: " + str(np.median(data["Potential"])))
print("Media Values: " + str(np.median(intValues)))
print("Media Wages: " + str(np.median(intWages)))
```

```python
import matplotlib.pyplot as plt

x = ['Overrall', 'Potential']
y = [np.median(data["Overall"]), np.median(data["Potential"])]

plt.title('MEDIA Overrall - Potential')

plt.bar(x, y)

plt.show()
```

```python
x = ['Values', 'Wages']
y = [np.median(intValues), np.median(intWages)]

plt.title('MEDIA VALUES - WAGES')

plt.bar(x, y)

plt.show()
```

```python
print("Desviación Típica Overrall: " + str(np.std(data["Overall"])))
print("Desviación Típica Potential: " + str(np.std(data["Potential"])))
print("Desviación Típica Values: " + str(np.std(intValues)))
print("Desviación Típica Wages: " + str(np.std(intWages)))
```

```python
print("Desviación Típica Por Columna: \n" + str(np.std(data, 0)))
```

```python
print("Varianza Overrall: " + str(np.var(data["Overall"])))
print("Varianza Potential: " + str(np.var(data["Potential"])))
print("Varianza Values: " + str(np.var(intValues)))
print("Varianza Wages: " + str(np.var(intWages)))
```

```python
print("Varianza Por Columna: \n" + str(np.var(data, 0)))
```

```python
corrOvVa = np.corrcoef(data["Overall"], intValues)
corrOvWa = np.corrcoef(data["Overall"], intWages)
corrPoVa = np.corrcoef(data["Potential"], intValues)
corrPoWa = np.corrcoef(data["Potential"], intWages)
corrValues = np.corrcoef(data["Overall"], data["Potential"], intValues)
corrWages = np.corrcoef(data["Overall"], data["Potential"], intWages)
```

```python
print("Correlación Overrall - Values: \n" + str(corrOvVa))
print("Correlación Overrall - Wages: \n" + str(corrOvWa))
print("Correlación Potential - Values: \n" + str(corrPoVa))
print("Correlación Potential - Wages: \n" + str(corrPoWa))
print("Correlación Overrall - Potential - Values: \n" + str(corrValues))
print("Correlación Overrall - Potential - Wages: \n" + str(corrWages))
```

```python
covOvVa = np.cov(data["Overall"], intValues)
covOvWa = np.cov(data["Overall"], intWages)
covPoVa = np.cov(data["Potential"], intValues)
covPoWa = np.cov(data["Potential"], intWages)
covValues = np.cov(data["Overall"], data["Potential"], intValues)
covWages = np.cov(data["Overall"], data["Potential"], intWages)

print("Covarianza Overrall - Values: \n" + str(covOvVa))
print("Covarianza Overrall - Wages: \n" + str(covOvWa))
print("Covarianza Potential - Values: \n" + str(covPoVa))
print("Covarianza Potential - Wages: \n" + str(covPoWa))
print("Covarianza Overrall - Potential - Values: \n" + str(covValues))
print("Covarianza Overrall - Potential - Wages: \n" + str(covWages))
```

```python
import seaborn as sns

df = pd.DataFrame({
    "Overrall": data["Overall"],
    "Potential": data["Potential"],
    "Value": intValues,
    "Wage": intWages})

print("Correlación")
sns.heatmap(df.corr())
```
