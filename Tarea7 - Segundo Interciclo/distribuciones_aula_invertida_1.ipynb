```python
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np
```

#Distribuciones Discretas

## 1. Distribución Hipergeométrica

### 1.1 Introducción:

La distribución hipergeométrica viene dada por la siguiente formula:

$$p(r; n, N, M) = \frac{\left(\begin{array}{c} M \\ r \end{array}\right)\left(\begin{array}{c} N - M \\ n - r \end{array}\right)}{ \left(\begin{array}{c} N \\ n \end{array}\right)}
$$

Dónde: 

a) r esta limitado por max(0, n - N + M) y min(n, M)

b) n esta limitado por (1 <= n <= N)

c) N esta limitado por (N >= 1)

d) M esta limitado por (M >= 1)

### 1.2 Descripción:

La distribución hipergeométrica describe experimentos donde un evento no puede ser seleccionado más de una vez o no tiene remplazo, por lo que la probabilidad de que un elemento sea seleccionado aumenta con cada experimentación.

La distribucion hipergeométrica puede ser utilizada por ejemplo para conocer, en un grupo de 10 personas la probabilidad de que la primera persona seleccionada tenga sangre O+ conociendo con aterioridad que 7 de las 10 cuentan con este tipo de sangre. Para la primera experimentación la probabilidad sera de 0.7000 pero para la segunda experimentación la probabilidad bajara al 0.66667. Esta variación puede aumentar a medida que el tamaña de la muestra incrementa.

### 1.3 Ejemplo de uso:

```python
# PARAMETROS DE FORMA
M, n, N = 30, 10, 12

# DISTRIBUCIÓN HIPERGEOMETRICA
hipergeometrica = stats.hypergeom(M, n, N)

x = np.arange(0, n + 1)

# FUNCIÓN DE MASA DE PROBABILIDAD
fmp = hipergeometrica.pmf(x)

# GRÁFICA
plt.plot(x, fmp, '--')
plt.vlines(x, 0, fmp, colors='b', lw=5, alpha=0.5)
plt.title('Distribución Hipergeométrica')
plt.ylabel('Probabilidad')
plt.xlabel('Valores')
plt.show()
```

```python
# GENERACIÓN DE ALEATORIOS
aleatorios = hipergeometrica.rvs(100)
cuenta, cajas, ignorar = plt.hist(aleatorios, 20)

# GRÁFICA HISTOGRAMA
plt.ylabel('Frecuencia')
plt.xlabel('Valores')
plt.title('Histograma Distribución Hipergeométrica')
plt.show()
```

```python
print(aleatorios)
```

## 2. Distribución Uniforme

### 2.1 Introducción:

La distribución uniforme viene dada por la siguiente formula:

$$p(x; a, b) = \frac{1}{b-a}$$

Donde:

a) a =< x <= b

b) p(x; a, b) = 0 Si x <= a

c) p(x; a, b) = x - a / b - a Si a <= x <= b

d) p(x; a, b) = 1 Si b <= x

### 2.2 Descripción:

La distribución uniforme es utilizada para describir que todas las variables definidas dentro de determinado rango, tiene todas la mismo probabilidad de ocurrencia. Es conocida también como distribución rectangular.

Aunque en la naturaleza no nos encontramos comunmente con este tipo de comportamiento, es util cuando se trabaja con procesos de simulación cuyos eventos cuentan con un grao de aparición regular. Por ejemplo cuando se intenta determinar el nivel de voltaje que llega a determinado hogar en un rango de 123.0 y 125.0 V. 

### 2.3 Ejemplo de Uso:

```python
# DISTRIBUCIÓN UNIFORME
uniforme = stats.uniform()

x = np.linspace(uniforme.ppf(0.01), uniforme.ppf(0.99), 100)

# FUNCIÓN DE PROBABILIDAD
fp = uniforme.pdf(x)

# GRÁFICO
fig, ax = plt.subplots()
ax.plot(x, fp, '--')
ax.vlines(x, 0, fp, colors='b', lw=5, alpha=0.5)
ax.set_yticks([0., 0.2, 0.4, 0.6, 0.8, 1., 1.2])
plt.title('Distribución Uniforme')
plt.ylabel('Probabilidad')
plt.xlabel('Valores')
plt.show()
```

```python
# GENERACIÓN DE ALEATORIOS
aleatorios = uniforme.rvs(100)
cuenta, cajas, ignorar = plt.hist(aleatorios, 20)

# GRÁFICA HISTOGRAMA
plt.ylabel('Frecuencia')
plt.xlabel('Valores')
plt.title('Histograma Uniforme')
plt.show()
```

```python
print(aleatorios)
```

## 3. Distribución Binomial Negativa

### 3.1 Introducción:

La distribución binomial negativa viene dada por:

$$p(X = k) = \frac{k - 1}{r - 1}p^r(1 - p)^{k-r}$$

Donde:

a) Cada secuencia es independiente de las otras

b) En cada intento unicamente se puede obtener 2 resultados (exito - fracaso)

c) La probabilidad de éxito es constante en cada secuencia

d) Los intentos continuan hasta que se consiga r éxitos

### 3.2 Descripción:

La distribución binomial negativa es una distribución que nos permite trabajar con modelos en los que se reitera determinada experimentación hasta conseguir un número exacto de resultados exitosos.

Un claro ejemplo de utilización de la distribución binomial negativa es cuando, al jugar futbol y proponernos lanzar tiros libres, no dejar de intentarlo hasta conseguir 10 goles en esta modalidad. Donde x sera p(x) sera el número de experimentos realizados hasta conseguir los 10 goles de tiro libre con anterioridad proyectados.

### 3.3 Ejemplo de uso:

```python
# PARAMETROS DE FORMA
n, p = 5, 0.5

# DISTRIBUCIÓN BINOMIAL NEGATIVA
bimNegativa = stats.nbinom(n, p)

x = np.arange(0, n + 1)

# FUNCIÓN DE MASA DE PROBABILIDAD
fmp = bimNegativa.pmf(x)

# GRÁFICA
plt.plot(x, fmp, '--')
plt.vlines(x, 0, fmp, colors='b', lw=5, alpha=0.5)
plt.title('Distribución Binomial Negativa')
plt.ylabel('Probabilidad')
plt.xlabel('Valores')
plt.show()
```

```python
# GENERACIÓN DE ALEATORIOS
aleatorios = bimNegativa.rvs(100)
cuenta, cajas, ignorar = plt.hist(aleatorios, 20)

# GRÁFICA HISTOGRAMA
plt.ylabel('Frecuencia')
plt.xlabel('Valores')
plt.title('Histograma Binomial Negativa')
plt.show()
```

```python
print(aleatorios)
```
